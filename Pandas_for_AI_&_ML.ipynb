Updating Rows and Columns - Modifying Data Within DataFrame

```python
df
```

```python
df.columns
```

```python
df.columns = [x.upper() for x in df.columns]    # changing columns name to upper case using list comprehension
df
```

```python
df.columns = [x.lower() for x in df.columns]    # changing columns name to lower case using list comprehension
df
```

```python
df.columns = df.columns.str.replace('_', ' ')   # replacing the columns names haveing '_' with ' '
df
```

```python
df.columns = df.columns.str.replace(' ', '_')   # replacing the columns names haveing ' ' with '_'
df
```

```python
# updating the first_name and email only using loc
df.loc[0, ['first_name', 'email']] = ['Nishan', 'nishan@gmail.xon']

df
```

```python
# updating rows using at
df.at[0, 'email'] = 'nishan@gmail.com'
df
```

```python
# using different methods to update rows
"""
1. apply
2. map
3. applymap
4. replace
"""
```

```python
# 1. apply
df['email'].apply(len)
```

```python
def update_email(email):
    return email.upper()

df['email'] = df['email'].apply(update_email)
df
```

```python
# using lambda function with apply method
df['email'] = df['email'].apply(lambda x: x.lower())
df
```

```python
# 2. applymap

df.applymap(len)
```

```python
df.applymap(str.lower)
```

```python
df
```

```python
# 3. map
df['first_name'].map({'Misam':'Anu', })
```

```python
# 4. replace
df['first_name'] = df['first_name'].replace({'Misam': 'Anu'})
```
